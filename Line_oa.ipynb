```python
def sort(nums: list[int]):
    count = 0
    for i in range(len(nums)):
        for j in range(len(nums)-1, i, -1):
            if nums[j] > nums[j-1]:
                nums[j], nums[j-1] = nums[j], nums[j-1]
                count += 1
    print(count)          
```

```python
arr = [1, -5, 6, 8]

sort(arr)
```

```python
def find_x(x: int):
    x |= x >> 1
    x |= x >> 2
    x |= x >> 4
    x |= x >> 8
    x |= x >> 16

    return (x + 1) >> 1

def solution(A):
    # write your code in Python 3.6

    # Calculate the binarian of input list and store it into sum
    sum = 0

    for i in range(len(A)):
        sum += 2 ** A[i]

    # Use a count variable to store the iteration times of subtraction
    count = 0

    while sum > 0:
        sum -= find_x(sum)
        count += 1
        
    print(count)
```

```python
A = [1,5,4,7]
solution(A)
```
